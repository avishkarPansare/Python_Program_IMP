```python
#create new db 
mng_db = mng_client['Admission_Prediction12'] 
```

```python
# create table  
collection_name = 'AdmissionPrediction343'
db_cm = mng_db[collection_name]
```

```python
# reading csv file from currnt dir
data = pd.read_csv('Admission_Prediction.csv')
```

```python
# canvert to csv reader object into json format 
data_json = json.loads(data.to_json(orient='records'))
# data_json
```

```python
# inserting data 
db_cm.insert(data_json)
print(mng_client.list_database_names())
```

```python
# printing only the five rows
result_total= db_cm.find({}).limit(5)
for res in result_total:
    print(res)
```

```python
# printing only the five rows
result_total= db_cm.find({}).skip(495)
for res in result_total:
    print(res)
```

```python
# retrieveing some columns
result_some= db_cm.find({}, {'GRE Score','TOEFL Score'}).skip(495) # retrieveing two columns
# The second parameter in find() specifies which columns to choose
for res in result_some:
    print(res)
```

```python
my_db_query={'GRE Score':{'$gt':331}}
# printing all the rows where GRE Score is greaterthan 330
result_GRE= db_cm.find(my_db_query)
for res in result_GRE:
    print(res)
```

```python
# finding the rows satisfying a given criteria
my_db_query={'GRE Score':'337'}
# printing all the rows where GRE Score is 337
result_GRE= db_cm.find(my_db_query)
for i in result_GRE:
    print(i)
```

```python
# we can sort the above results as well, this has been sorted in ascending order
my_db_query={'GRE Score':337}
result_GRE= db_cm.find(my_db_query,{'GRE Score','TOEFL Score'}).sort('GRE Score')
for res in result_GRE:
    print(res)
```

```python
# we can sort the above results as well, this has been sorted in descending order
my_db_query={'GRE Score':{'$gt':330}}
result_GRE= db_cm.find(my_db_query,{'GRE Score','TOEFL Score'}).sort('GRE Score',-1)
for res in result_GRE:
    print(res)
```

```python
# Deleting records from mongo DB:
# finding the rows satisfying a given criteria
my_db_query={'GRE Score':'337'}
x=db_cm.delete_one(my_db_query) # the deletion step
print(x.deleted_count)
```

```python
# updating the values present in collection

update_criteria={'GRE Score':340}
new_values= { "$set": { "GRE Score": 350 } }

x = db_cm.update_many(update_criteria, new_values)

print(x.modified_count)
```
